```python
class Rectangle:
  def __init__(self,length,width):
    self.length=length
    self.width=width
  def Perimeter(self):
    return (self.length+self.width)*2
  def Area(self):
    return self.length*self.width
  def display(self):
    print("The length of rectangle is: ",self.length)
    print("The width  of rectangle is: ",self.width)
    print("The perimeter of rectangle is: ",self.Perimeter())
    print("The area of rectangle is: ",self.Area())

my_Rectangle=Rectangle(3,4)
my_Rectangle.display()
```

```python
import math
class Computation:
  def __init__(self):
     pass

  def factorial(self,n):
    s=1
    for i in range(1,n+1):
      s*=i
    print(f"Factorial is : {s}")

  def naturalSum(self,n):
    s=0
    for i in range(n+1):
      s+=i
    print(f"Sum of natural number : {s}")

  def primeNum(self, n):
    if n <= 1:
        print("Number is not prime")
        return

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            print("Number is not prime")
            return

    print("Number is prime")

  def testprims(self,a,b):
    if math.gcd(a,b)==1:
      print(f"{a} and {b} are prime to each other")
    else:
      print(f"{a} and {b} are not prime to each other")
  def tableMult(self,n):
     for i in range(1,11):
        print(i*n)
  def allTablesMult(self,*args):
     for i in args:
        for j in range(1,11):
           print(j*i)

  def listDiv(self,n):
    li=[]
    for i in range(1,n+1):
      if n%i==0:
        li.append(i)
    print(li)

  def fun_primeNum(self, n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

  def listDivPrim(self,n):
    lis=[]
    for i in range(1,n+1):
      if n%i==0 and self.fun_primeNum(i):
          lis.append(i)
    print(lis)


b_fun=Computation()
b_fun.factorial(5)
b_fun.naturalSum(10)
b_fun.primeNum(11)
b_fun.testprims(4,9)
b_fun.tableMult(2)
b_fun.allTablesMult(3,4,5)
b_fun.listDiv(12)
b_fun.listDivPrim(12)
```

```python
class Complex:
  def __init__(self,real,img):
    self.real=real
    self.img=img
  def showNumber(self):
    print(self.real,"i +",self.img,"j")
  def __add__(self,num2):
    newReal=self.real+num2.real
    newImg=self.img+num2.img
    return Complex(newReal,newImg)

num1=Complex(1,3)
num1.showNumber()
num2=Complex(4,6)
num2.showNumber()
num3=num1+num2
num3.showNumber()
```

```python
class Instructor:
    def __init__(self):
        self.__name = None
        self.__technology_skill = []
        self.__experience = 0
        self.__avg_feedback = 0.0

    # Setter methods
    def set_name(self, name):
        self.__name = name

    def set_technology_skill(self, skills):
        self.__technology_skill = skills  # skills should be a list

    def set_experience(self, experience):
        self.__experience = experience

    def set_avg_feedback(self, avg_feedback):
        self.__avg_feedback = avg_feedback

    # Getter methods (optional, if needed)
    def get_name(self):
        return self.__name

    def get_technology_skill(self):
        return self.__technology_skill

    def get_experience(self):
        return self.__experience

    def get_avg_feedback(self):
        return self.__avg_feedback

    # Public method to check eligibility
    def check_eligibility(self):
        if self.__experience > 3 and self.__avg_feedback >= 4.5:
            return True
        elif self.__experience <= 3 and self.__avg_feedback >= 4.0:
            return True
        else:
            return False

    # Public method to allocate course
    def allocate_course(self, technology):
        if self.check_eligibility() and technology in self.__technology_skill:
            return True
        else:
            return False


# ------------------ Testing the Class ------------------

# Creating first instructor
instructor1 = Instructor()
instructor1.set_name("Alice")
instructor1.set_experience(5)
instructor1.set_avg_feedback(4.6)
instructor1.set_technology_skill(["Python", "Java", "C++"])

print("Instructor 1 Eligible:", instructor1.check_eligibility())  # True
print("Can teach Python:", instructor1.allocate_course("Python"))  # True
print("Can teach C#", instructor1.allocate_course("C#"))  # False

# Creating second instructor
instructor2 = Instructor()
instructor2.set_name("Bob")
instructor2.set_experience(2)
instructor2.set_avg_feedback(3.9)
instructor2.set_technology_skill(["HTML", "CSS", "JavaScript"])

print("\nInstructor 2 Eligible:", instructor2.check_eligibility())  # False
print("Can teach JavaScript:", instructor2.allocate_course("JavaScript"))  # False

# Creating third instructor
instructor3 = Instructor()
instructor3.set_name("Carol")
instructor3.set_experience(3)
instructor3.set_avg_feedback(4.2)
instructor3.set_technology_skill(["SQL", "Python"])

print("\nInstructor 3 Eligible:", instructor3.check_eligibility())  # True
print("Can teach Python:", instructor3.allocate_course("Python"))  # True
```
