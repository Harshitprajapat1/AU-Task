```python
#Q-3Write a function to create a 2d array with 1 on the border and 0 inside. Take 2-D array
# shape as (a,b) as parameter to function.
c=int(input("Enter row "))
d=int(input("Enter column "))
ma=np.ones((c,d))
for i in range(1,c-1):
  for j in range(1,d-1):
    ma[i][j]=0

print(ma)
```

```python
#Q-7: Consider a random integer (in range 1 to 100) vector with shape (10,2) representing
# coordinates, and coordinates of a point as array is given. Create an array of distance of each
# point in the random vectros from the given point. Distance array should be interger type.
t=np.random.randint(low=1,high=101,size=(10,2))
print(t)
gp = np.array([100,100])
distances = np.sqrt(np.sum((t - gp) ** 2, axis=1))
distances.astype("int")
```

```python
###Q-8: Consider a (6,7,8) shape array, what is the index (x,y,z) of the 100th element?
kl=np.arange(1,337).reshape(6,7,8)
# kl[1][5][3]
index=np.where(kl==100)
print(index)
```
