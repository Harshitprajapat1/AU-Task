```python
# Problem 5: Write a Python Program to Find the Sum of the Series till the
# nth term:
# 1 + x^2/2 + x^3/3 + â€¦ x^n/n n will be provided by the user
n=int(input("Enter nth term of number : "))
x=int(input("Enter  x  term of number : "))
sum=1
for i in range(2,n+1):
  if sum==1:
    print(f"{sum}+x^2/2",end="")
  else:
    print(f"+x^{i}/{i}",end="")
  sum+=(x**i)//i
print("\nSum of series is : ",sum)
```

```python
#Problem 10: Write a program that will take 2 numbers as input and prints the LCM and
# HCF of those 2 numbers
num1=int(input("Enter first number : "))
num2=int(input("Enter second number : "))
#Hcf
a=num1
b=num2
r=1
while r>0:
  r=a%b
  a=b
  b=r
print("HCF is : ",a)
#LCM
lcm=(num1*num2)//a
print("LCM is : ",lcm)
```

```python
# Problem 13:Given string contains a combination of the lower and upper
# case letters. Write a program to arrange the characters of a string so
# that all lowercase letters should come first.
# Given:
# str1 = PyNaTive
# Expected Output:
# yaivePNT
st1="PyNaTive"
low=''
upp=''
for i in st1:
  if i.islower():
    low+=i
  else:
    upp+=i
st2=low+upp
print(st2)
```

```python
# Problem 14:Take a alphanumeric string input and print the sum and
# average of the digits that appear in the string, ignoring all other
# characters.
# Input:
# hel123O4every093
# Output:
# Sum: 22
# Avg: 2.75

st23="hel123O4every093"
s=0
c=0
for i in st23:
  if i.isdigit():
    s+=int(i)
    c+=1
print("Sum : ",s)
print("Avg : ",s/c)
```

```python
# Problem 16: Check whether the string is Symmetrical.
# Statement: Given a string. the task is to check if the string is symmetrical or not. A string is said
# to be symmetrical if both the halves of the string are the same.
# Example 1:
# Input
# khokho
# Output
# The entered string is symmetrical
ss=input("Enter a word : ")
mid=len(ss)//2
if ss[:mid]==ss[mid:]:
  print("The string is symmetrical")
else:
  print("Srting is not symmetrical")
```

```python
# Problem 17: Reverse words in a given String
# Statement: We are given a string and we need to reverse words of a given string.
# Example 1:
# Input:
# geeks quiz practice code
# Output:
# code practice quiz geeks

sw="geeks quiz practice code"
word=sw.split()
reverse=word[::-1]
print(reverse)
```
